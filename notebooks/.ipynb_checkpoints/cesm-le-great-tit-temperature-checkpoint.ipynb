```python
%matplotlib inline
import os
import shutil

import numpy as np
import xarray as xr

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from cartopy.util import add_cyclic_point

import intake
```

```python
from ncar_jobqueue import NCARCluster
from dask.distributed import Client
try:
    cluster
    client
except:
    cluster = NCARCluster(memory='100 GB')
    cluster.scale(32)
    client = Client(cluster) # Connect this local process to remote workers'
client
```

```python
catalog_file = 'glade-cesm1-le.json'
col = intake.open_esm_datastore(catalog_file, sep='|')
col
```

```python
%%time
col_sub = col.search(experiment=['20C', 'RCP85'],                      
                     variable=['TREFHT'],
                     stream='cam.h1',
                     member_id=1
                    )
dsets_001 = col_sub.to_dataset_dict(cdf_kwargs={'chunks': {'time': 365*5}, 'decode_times': True})
_, ds_001 = dsets_001.popitem()
ds_001
```

```python
def subset(ds):
    ds['lon'] = ds_001.lon
    ds['lat'] = ds_001.lat
    return (ds[['TREFHT', 'lon', 'lat', 'time', 'time_bnds']]
            .sel(
                lat=slice(45, 60), 
                lon=slice(0, 15),
                time=slice('1920', '2100'),
            )
           )
```

```python
col_sub = col.search(experiment=['20C', 'RCP85'],                      
                     variable=['TREFHT'],
                     stream='cam.h1',
                    )
col_sub.df
```

```python
%%time
dsets = col_sub.to_dataset_dict(cdf_kwargs={'chunks': {'time': 365*5}, 'decode_times': True},
                               preprocess=subset)
dsets
```

```python
%%time
ordered_dsets_keys = ['atm|20C|cam.h1', 'atm|RCP85|cam.h1']
ds = xr.concat(
    [dsets[exp] for exp in ordered_dsets_keys], 
    dim='time', 
    data_vars='minimal'
).compute()
ds
```

```python
lon, lat = 5.85573, 52.0407254

def plot_field(da):
    import cartopy.feature as cfeature
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.EuroPP())
    da.plot(
        ax=ax,
        transform=ccrs.PlateCarree(),        
    )
    ax.coastlines()
    ax.add_feature(cfeature.BORDERS, linestyle='-')
    ax.plot(lon, lat, 'r*', transform=ccrs.PlateCarree())

plot_field(ds.TREFHT.sel(time=slice('1920', '1960')).mean(['member_id', 'time']))
plt.title('Mean (1920-1960)')

plot_field(ds.TREFHT.sel(time=slice('1920', '1960')).std(['member_id', 'time']))
plt.title('Std. Dev. (1920-1960)')
```

```python
%%time
dsp = ds.interp(lat=lat, lon=lon).reset_coords(['lat', 'lon'], drop=True)
dsp
```

```python
dsp_mon = dsp.resample(time='M').mean()
dsp_mon
```

```python
for member_id in dsp.member_id.values:
    dsp_mon.TREFHT.sel(member_id=member_id).plot()
    
```

```python
for member_id in dsp.member_id.values:
    dsp.TREFHT.sel(member_id=member_id).to_dataframe().to_csv(f'data/cesm-le-20C-RCP85.TREFHT.{member_id:03d}.csv')
```
